```python
num_features = df.select_dtypes(exclude='O').columns
cat_features = df.select_dtypes('O').columns
```

```python
num_features
```

```python
for feature in num_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
for i in num_features:
    sns.barplot(df[i])
    plt.show()
```

```python
# sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=df, alpha=0.6, s=50)

# plt.show()

for i in num_features:
    sns.scatterplot(x=df[i],y='LoanAmount',data=df, alpha=0.6, s=50)
    plt.show()
```

```python
for i in num_features:
    sns.lineplot(df[i])
    plt.show()
```

```python
for i in num_features:
    sns.histplot(df[i])
    plt.show()
```

```python
corr_matrix=df.select_dtypes(exclude='O').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
